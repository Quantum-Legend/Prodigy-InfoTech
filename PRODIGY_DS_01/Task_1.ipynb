# Task 1
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of genders and age in a population.

```python
# Import the libraries
import seaborn as sns # for reading the data
import matplotlib.pyplot as plt # for creating charts
```

### Loading the dataset
Let us consider the "Iris" dataset for this project. We will load the iris dataset using the seaborn library.

```python
# Load the iris dataset using seaborn
iris_dataset = sns.load_dataset("iris")
```

```python
iris_dataset.head()
```

### Histogram for plotting a Continuous Variable
The iris dataset has continuous variables that contain information about sepal and petal measurements for different species of iris flowers.
Let us plot the distribution of the sepal length using a histogram. 

```python
# Plotting the distribution of sepal length
plt.figure(figsize=(12, 6))
sns.histplot(iris_dataset['sepal_length'], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.show()
```

### Bar Chart for plotting a Categorical Variable
The Species variable in the iris dataset is a categorical variable. The iris is classified in three categories of species - setosa, versicolor and virginica. Let us plot their distribution using a bar chart.

```python
# Plotting the distribution of species
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris_dataset, palette='pastel', edgecolor='black')
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
